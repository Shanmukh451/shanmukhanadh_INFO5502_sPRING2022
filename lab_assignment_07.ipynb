```python
### You code here


import pandas as pd

fd = pd.read_csv("/content/drive/MyDrive/portfolio_data.csv")
fd.head(10)




```

```python
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
plt.style.use("fivethirtyeight")
%matplotlib inline
from sklearn.model_selection import train_test_split as Split
from sklearn.linear_model import LinearRegression as LR
from sklearn.metrics import mean_squared_error, r2_score

a = fd.drop(['Date', 'AMZN'],axis=1)
b = fd['AMZN']
atrain,atest,btrain,btest = Split(a, b, test_size = 0.35, random_state = 0)
lom = LR()
lom.fit(atrain, btrain)

yprediction = lom.predict(atest)
print("RMSE value : ",np.sqrt(mean_squared_error(btest, yprediction)))
print("R2 value ", r2_score(btest, yprediction))
```

```python
from sklearn.ensemble import RandomForestRegressor

xtrain, xtest, ytrain, ytest = train_test_split(a, b, test_size = 0.3, random_state = 0)
rfreg = RandomForestRegressor(random_state=211)
rfreg.fit(xtrain, ytrain)

r = rfreg.predict(xtest)
print("RMSE value",np.sqrt(mean_squared_error(ytest, r)))
print("R2 value", r2_score(ytest, r))
```

```python
def process_tweet(tweet):
    return " ".join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])", " ",tweet.lower()).split())
```

```python
fd['Cleaned_tweets'] = df['tweet'].apply(process_tweet)
fd.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.linear_model import LogisticRegression

cnt_non_fraud = fd[fd['label'] == 0]['processed_tweets'].count()
fd_class_fraud = fd[fd['label'] == 1]
fd_class_nonfraud = fd[fd['label'] == 0]
fd_class_fraud_oversample = fd_class_fraud.sample(cnt_non_fraud, replace=True)
fd_oversampled = pd.concat([fd_class_nonfraud, fd_class_fraud_oversample], axis=0)

print('Total No of Random over-sampling by values:')
print(fd_oversampled['label'].value_counts())

a = fd_oversampled['processed_tweets']
b = fd_oversampled['label']

A_train, B_test, a_train, b_test = train_test_split(a, b, test_size=0.35, stratify = None)

c_v = CountVectorizer(stop_words='english')
transformer = TfidfTransformer(norm='l2',sublinear_tf=True)
A_train_counts = c_v.fit_transform(A_train)
a_train_tfidf = transformer.fit_transform(A_train_counts)
b_test_counts = c_v.transform(B_test)
b_test_tfidf = transformer.transform(b_test_counts)

logregression = LogisticRegression(random_state=0)
logregression.fit(a_train_tfidf,a_train)
prediction_logs = logregression.predict(b_test_tfidf)
print(confusion_matrix(b_test,prediction_logs))
print(classification_report(b_test, prediction_logs))
```
